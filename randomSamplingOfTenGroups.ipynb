```python
#number of times an experiment should be repeated for a 95% probability

import numpy as np
import pandas as pd
import random
import math
```

```python
N = 10
P = 0.95
n = math.ceil(math.log(1 - P) / math.log(1 - 1/N))
print(f" {n} times")
```
